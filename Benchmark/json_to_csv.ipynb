```python
import json
import pandas as pd

json_data_path = "/content/MaSaC_train_erc.json"

with open(json_data_path, 'r') as json_file:
    data = json.load(json_file)

episode_numbers = []
speakers = []
utterances = []
emotions = []
emotion_name = []

emotion_mapping = {
    "neutral": 0,
    "contempt": 1,
    "anger": 2,
    "disgust": 3,
    "joy": 4,
    "sadness": 5,
    "fear": 6,
    "surprise": 7
}

episode_counter = 0

for episode_data in data:
    episode_length = len(episode_data["utterances"])

    episode_numbers.extend([episode_counter] * episode_length)

    speakers.extend(episode_data["speakers"])
    utterances.extend(episode_data["utterances"])
    emotion_name.extend(episode_data["emotions"])
    emotions.extend([emotion_mapping[emotion] for emotion in episode_data["emotions"]])

    episode_counter += 1

    episode_numbers.append(None)
    speakers.append(None)
    utterances.append(None)
    emotions.append(None)
    emotion_name.append(None)

episode_df = pd.DataFrame({
    "Dialogue_Id": episode_numbers,
    "Speaker": speakers,
    "Emotion": emotions,
    "Emotion_name": emotion_name,
    "Utterance": utterances,
})

csv_file_path = "MaSaC_train_erc.csv"
episode_df.to_csv(csv_file_path, index=False)

print(f"CSV file saved to {csv_file_path}")
```
